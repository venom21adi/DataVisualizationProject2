### For better understanding we are mentioning what each column title actually means.

<b>Pregnancies</b> - Number of pregnancies the person had<br>
<b>Glucose</b> - Glucose level in the blood at the time of meausrement
(Plasma glucose concentration a 2 hours in an oral glucose tolerance test)<br>
<b>Blood Pressure</b> - Blood Pressure level in the blood at the time of meausrement<br>
<b>Skin Thickness</b> - Skin Thickness at the time of measurement
(Triceps skin fold thickness (mm))<br>
<b>Insulin</b> - Insulin level at the time of measurement<br>
(2-Hour serum insulin (mu U/ml))
<b>BMI</b> - BMI (Body Mass Index) of the person at the time of measurement<br>
<b>DiabetesPedigreeFunction</b> - Diabetes pedigree function (a function which scores likelihood of diabetes based on family history)<br>
<b>Age</b> - Age of the person<br>
<b>Outcome</b> - The person has diabetes or not
<br>
<ul>If the value is 1 then the person has diabetes.</ul>
<ul>If the value is 0 then the person doesn't have diabetes.</ul><br>

## Let's see how number of pregnancies vary

### It is a categorical variable so we can use a bar plot to visualize it.

### In order to do so we need to extract the patients based on their number of pregnancies  

### For example, there may be 200 patients with 0 pregnancies 300 with 1 and so on.

### To peform that we will use the value_counts function of the pandas dataframe and store that value seperately and then use that for visualization.

```python
#Extracting the counts of pregnancies
data1 = dict(df['Pregnancies'].value_counts())
```

```python
data1
```

```python
#Let's use the bar plot
#Setting the figure size
plt.figure(figsize=(16,12))
plt.bar(x = data1.keys(),height=data1.values())
plt.title("Number of patientes based on the number pregnancies")
plt.xticks(list(data1.keys()))
plt.xlabel("Pregnancies")
plt.tight_layout()
plt.ylabel("Number of patientes")
```

## Conclusion

#### As we can see from the above graph that most of the people have 0,1 and 2 number of pregnanices.

#### ------------------------------------------------------------------------------------------------------------------------------------------------------------

## Let's see how glucose varies

### It is a continuous variable so we can use a line plot to visualize it.

```python
#Let's use the line plot
#Setting the figure size
plt.figure(figsize=(15,9))
plt.plot(df['Glucose'],label = 'Glucose')
plt.title("Glucose")
plt.xlabel("Number of Patients")
plt.ylabel("Value of blood glucose level")
plt.legend()
```

## Conclusion

#### From the above plot, it can be understood that there is a significant amount of variation in the glucose level.
#### However, this is not enough to understand the data
#### We need to do more investigation let's plot the mean, maximum and the minimum line.

```python
#Let's see how glucose level varies using line plot
#Setting the figure size
plt.figure(figsize=(11,6))
plt.plot(df['Glucose'])
plt.title("Glucose")
plt.hlines(y = df['Glucose'].max(),label = "Maximum blood glucose level",xmin =0,xmax = 800,color ='green')
plt.hlines(y = df['Glucose'].mean(),label = "Average blood glucose level",xmin =0,xmax = 800,color ='black')
plt.hlines(y = df['Glucose'].min(),label = "Minmum blood glucose level",xmin =0,xmax = 800,color ='red')
plt.xlabel("Number of Patients")
plt.ylabel("Value of blood glucose level")
plt.legend()
```

## Conclusion
#### From the above plot, we can understand that only a few entries are very far from the mean line.
#### Those entries which are very far from the mean can be considered as outliers.
#### However, there are specific methods to find the outliers which are discussed in the Applied Statistics section.

## Let's see how blood pressure varies

### It is a continuous variable so we can use a histogram plot to visualize it.

### This is a diastolic blood pressure which is often reffered as the lower blood pressure value

### For example, if a patient has a 120/80 blood pressure then 80 is the diastolic blood pressure value.

```python
#Let's see how blood pressure level varies using blood pressure plot
#Setting the figure size
plt.figure(figsize=(11,6))
plt.hist(x =df["BloodPressure"])
plt.title("Diastolic Blood Pressure")
plt.xlabel("Value of Blood Pressure (mm Hg)")
plt.ylabel("Number of Patients")
```

## Conclusion
#### From the above plot, maximum people have Diastolic blood pressure in the range of 60-80.
#### Further, there are few patients which can be considered as outliers and have extreme values such as above 100 or 0.
#### As we know the value of Diastolic blood pressure can not be 0 this indicates meausrement or recording errors.
#### Those entries which are very far from the mean can be considered as outliers.
#### However, there are specific methods to find the outliers which are discussed in the Applied Statistics section.
